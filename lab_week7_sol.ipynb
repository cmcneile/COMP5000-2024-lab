##  Question 1b, Calculations with numpy

The python code below shows the exam and coursework marks for four students.


```python
import numpy as np

exam = np.array( [ 55 , 45 , 67 , 60   ]   ) 
coursework = np.array( [ 45 , 50 , 81, 34  ] ) 
```

The total mark for the module is 60% from the exam mark
and 40% from coursework mark. Write some python code to compute the
total mark for all four students and the mean of the total mark for
the students. See
https://scipy-lectures.org/intro/numpy/operations.html for some
ideas.

The output should like the below:
<pre>
Total marks for each student  [51.  47.  72.6 49.6]

```python
# Put your solution here

import numpy as np

exam = np.array( [ 55 , 45 , 67 , 60   ]   ) 
coursework = np.array( [ 45 , 50 , 81, 34  ] ) 

mark = 0.6 * exam + 0.4*coursework

print("Total marks for each student " , mark)
print("Class mean = " , np.mean(mark) , "%"  )
```

##  Question 1c, Working with multi-dimensional numpy arrays

Modify the python code below so that the script finds the maximum and minimum for each row.

```python
import numpy as np
arr= np.array([[1,23,78],[98,60,75],[79,25,48]])  
print(arr)

print("Entire array, minimum " , np.amin(arr),
   "Maximum ", np.amax(arr))
```

```python
# Put your solution here

#https://data-flair.training/blogs/numpy-statistical-functions/

import numpy as np
arr= np.array([[1,23,78],[98,60,75],[79,25,48]])  
print(arr)

print("Entire array, minimum " , np.amin(arr),
   "Maximum ", np.amax(arr))

for i in range(0,3) :
  print("Row " , i , " of array , minimum= " , np.amin(arr[i,:]),
   "maximum= ", np.amax(arr[i,:]))

```

```python
# Put your solution here
import numpy as np

a = np.array(42)
b = np.array([1, 2, 3, 4, 5])
c = np.array([[1, 2, 3], [4, 5, 6]])
d = np.array([[[1, 2, 3], [4, 5, 6]], [[-1, -2, -3], [-4, -5, -6]]])


def look_array(x_) :
   print(x_)
   print("Number of dimensions ", x_.ndim)
   print("Dimensions of the array " , x_.shape)
   print(" ")

look_array(a)
look_array(b)
look_array(c)
look_array(d)
```
